```python
import matplotlib.pyplot as plt

import numpy as np

# Criando os arrays (matrizes) para utilizar nos gráficos

x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

# Criando um gráfico com os valores anteriores
plt.plot(x,y)
```

MODIFICANDO 

```python
plt.plot(x, y, marker = 'o', c= 'red')
```

Modificando a forma como os pontos são ligados e o tipo de linha

```python
plt.plot(
    x,
    y,
    marker = 'o',
    c = 'blue',
    ds = 'steps', # Forma de ligação
    ls = '--' # Estilo de linha
)
```

Alterando as cores dos marcadores

```python
plt.plot(
    x,
    y,
    marker = 'o', # Marcador do gráfico
    c = 'red', # Cor da linha
    ds = 'steps', # Forma de ligação
    ls = '--', # Estilo de linha
    mec = 'blue' # Cor dos marcadores
)
```

Adicionando outros elementos ao gráfico


```python
plt.plot(x,y) # cria o gráfico

plt.xlabel('valor de x', fontsize = 12) # Adicionando o título do eixo x

plt.ylabel('Valores de y', fontsize = 12) # Adicionando o titulo eixo y

plt.title('Titulo do Gráfico', fontsize = 20) # Alterando as cores do rótulo de X e Y
plt.xticks(fontsize = 8, c = 'blue')
plt.xticks(fontsize = 8, c = 'purple')
```
